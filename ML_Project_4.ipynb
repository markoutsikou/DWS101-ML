# **Question 1 - Loading Dataset, Histograms**

```python
df = pd.read_csv('wine-missing.csv')
df.describe()
```

```python
df['fixed acidity'].hist(bins = 20, color = 'darkorange', edgecolor = 'black', grid = False)
plt.title('Fixed Acidity Histogram', fontweight = 'bold')
plt.xlabel('Fixed Acidity', fontweight = 'bold')
plt.ylabel('Counts', fontweight = 'bold')
plt.show()
```
